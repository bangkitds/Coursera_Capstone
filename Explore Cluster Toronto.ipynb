Get Data Latitude and Longitude and Merge it with Toronto Data based on Postal Code

```python
df_geo= pd.read_csv ('Geospatial_Coordinates.csv')
df_geo
```

```python
group= pd.merge(df1,df_geo,on='Postal Code')
group
```

Explore and cluster the neighborhoods in Toronto

```python
address = 'Toronto, TR'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(group['Latitude'], group['Longitude'], group['Borough'], group['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
